```python
plt.figure(figsize=(10, 6))
plt.bar(df["Item Type"],df["Total sale"],width= 0.7,color="black")
plt.title("Sale by Item Type",fontsize=20)
plt.xlabel("Item type",fontsize=12)
plt.ylabel("Sale",fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
Countit_sales = df.groupby('Item Type')['Total sale'].sum()
top5_isale = Countit_sales.sort_values(ascending=False).head(5)


# # Create the bar plot
plt.figure(figsize=(10, 6))
top5_isale.plot(kind='bar', color="Red")

# # Add titles and labels
plt.title('Top 5 Sales by ITEM TYPE', fontsize=16)
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
Countit_p = df.groupby('Item Type')['Total Profit'].sum()
top5_p = Countit_p.sort_values(ascending=False).head(5)


# # Create the bar plot
plt.figure(figsize=(10, 6))
top5_p.plot(kind='bar', color="green")

# # Add titles and labels
plt.title('Top 5 Profit by ITEM TYPE', fontsize=16)
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
Count = df.groupby('Country')['Total Cost'].sum()
top = Count.sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
top.plot(kind='line')

# # Add titles and labels
plt.title('Top 5 Cost by Country', fontsize=16)
plt.xlabel('Item Type',fontsize=10)
plt.ylabel('Cost',fontsize=10)
plt.show()
```

```python
Con_sales = df.groupby('Country')['Total sale'].sum()
top_5_sales = Con_sales.sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
top_5_sales.plot(kind='pie',autopct='%1.2f%%',explode=(0.2, 0, 0,0,0), shadow=True,colors=('red','blue','green','yellow','orange'))

# # Add titles and labels
plt.title('Top 5 Sales by Country', fontsize=16)
plt.ylabel('')
plt.show()
```

```python
Countit_saless = df.groupby('Country')['Total Profit'].sum()
top5_isalee = Countit_saless.sort_values(ascending=False).head(5)


# # Create the bar plot
plt.figure(figsize=(10, 6))
top5_isalee.plot(kind='pie',autopct='%1.2f%%',explode=(0.2, 0, 0,0,0), shadow=True,colors=('green','blue','red','yellow','orange'))

# # Add titles and labels
plt.title('Top 5 Profit by Country', fontsize=16)
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.barh(df["Sales Channel"],df["Total sale"],color="Yellow")
plt.title("Sale by Sales Channel",fontsize=20)
plt.xlabel("Sales",fontsize=12)
plt.ylabel("Sales Channel",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.barh(df["Sales Channel"],df["Total Profit"],color="blue")
plt.title("Profit by Sales Channel",fontsize=20)
plt.xlabel("Profit",fontsize=12)
plt.ylabel("Sales Channel",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df[['Total sale', 'Total Cost', 'Total Profit']],sym='',showmeans=True, meanline=True, vert=True, labels=['Total Sale', 'Total Cost', 'Total Profit'])

# Add labels and title
plt.xlabel('Variables')
plt.ylabel('Values')
plt.title('Boxplot of Total Sale, Total Cost, Total Profit')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# plt.hist(df["Item Type"], bins=12, color='yellow', edgecolor='black')
sns.histplot(data=df,x=df['Item Type'],y=df['Total sale'])

# Add labels and title
plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.xticks(rotation=70)

plt.title('Histogram')

# Show plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Item Type', y='Total Profit', data=df)
plt.xticks(rotation=45)
plt.title('Total Profit by Item Type')
plt.show()
```
